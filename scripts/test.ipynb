## Confidence Weights

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import zscore

def get_confweights():
    train_sig_path = '/data/leslie/shared/ASA/mouseASA/data/cd8/betabinom_result_combCounts_150bp_trainOnly.csv'
    pvals = pd.read_csv(train_sig_path)['p.adj']
    zscores = zscore(-np.log10(pvals+1e-5))
    zscores += abs(min(zscores))
    return zscores

plt.hist(get_confweights(), bins=50)
plt.show()
```

```python
temp = get_confweights()
len(temp[temp>0.5])
```
